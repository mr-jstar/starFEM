```julia
msh=loadMeditINRIA("tx.mesh")

```

```julia
#front
skin=1005

BelectrodeP = 1050

rad=0.01           # 12,12,12,10 mm
phi0,dphi=55,14.   # 45,18, 45,18, 45,18 ,50,17
the0,dthe=90,0.    # 25,45,90,90
dy=-0.039          #  0, 0, -0.006, -0.036
dz=0
n=5

lines = zeros(n,6)
for i=1:n
    lines[i,2] = dy
    lines[i,3] = dz
    lines[i,4:6] = [cosd(phi0+(i-1)*dphi) sind(phi0+(i-1)*dphi) sind(the0+(i-1)*dthe)]
end
```

```julia
#back
skin=1005

BelectrodeP = 1070

rad=0.01           # 12
phi0,dphi=225,18.   # 225,18,
the0,dthe=90,0.     # 0,20,45,67.5
dy=0                #  0, 
dz=0.009
n=5

lines = zeros(n,6)
for i=1:n
    lines[i,2] = dy
    lines[i,3] = dz
    lines[i,4:6] = [cosd(phi0+(i-1)*dphi) sind(phi0+(i-1)*dphi) sind(the0+(i-1)*dthe)]
end
```

```julia
lines
```

```julia
for k=1:n
    line = lines[k,1:3],lines[k,4:6]
    electrodeP = BelectrodeP + k
    faces = skinElectrode( msh, line, rad, 1, skin, electrodeP )
    msh["faces"] = vcat( msh["faces"], faces )
end
```

```julia
saveMeditINRIA("t1.mesh",msh)
```
